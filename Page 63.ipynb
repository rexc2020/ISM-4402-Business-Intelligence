```python
import pandas as pd
```

```python
Location = "gradedata.csv"
```

```python
df = pd.read_csv(Location)
```

```python
df.head()
```

```python
dummy = pd.get_dummies(df['gender'])
```

```python
dummy.head()
```

```python
df2 =pd.concat((df, dummy), axis=1)
```

```python
df2.head()
```

```python
df2 = df2.drop(["male"], axis=1)
```

```python
df2.head()
```

```python
df2.rename(columns={"female": "gender"})
```

```python
import statsmodels.formula.api as sm
```

```python
result = sm.ols(
    formula='grade ~ exercise + gender',
    data=df2).fit()
```

```python
result.summary()
```
